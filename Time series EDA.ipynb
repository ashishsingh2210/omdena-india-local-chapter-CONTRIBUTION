```python
interest_rate = pd.read_csv('Interest Rates RBI .csv',header=1)
```

```python
interest_rate.dropna(inplace=True)
```

```python
month  = interest_rate.Month.str.split('-',expand=True)[0].to_list()
```

```python
for i in range(len(month)):
    if month[i][0:2] == 'Ja':
        month[i] = "January"
    
    elif month[i][0] == 'F':
        month[i] = "February"
    
    elif month[i][0:3] == 'Mar':
        month[i] = "March"
    
    elif month[i][0:2] == 'Ap':
        month[i] = "April"
    
    elif month[i][0:3] == 'May':
        month[i] = "May"
    
    elif month[i][0:3] == 'Jun':
        month[i] = "June"
    
    elif month[i][0:3] == 'Jul':
        month[i] = "July"
    
    elif month[i][0:2] == 'Au':
        month[i] = "August"
    
    elif month[i][0:3] == 'Sep':
        month[i] = "September"
    
    elif month[i][0] == 'O':
        month[i] = "October"
    
    elif month[i][0] == 'N':
        month[i] = "November"
    
    elif month[i][0] == 'D':
        month[i] = "December"
```

```python
interest_rate['Date'] = pd.to_datetime({'year':interest_rate.Month.str.split('-',expand=True)[1],
               'month':[strptime(i,'%B').tm_mon for i in month],
               'day':np.ones(interest_rate.shape[0])})
```

```python
interest_rate.drop(['Month'],axis='columns',inplace=True)
```

```python
interest_rate.sort_values(by=['Date'],ascending=False,inplace=True)
```

```python
for i in interest_rate.columns:
    if i != 'Date':
        fig = px.line(x = interest_rate.Date, y = interest_rate[i],title=i)
        fig.update_xaxes(
                        rangeslider_visible=True,
                        )
        fig.show()
```
