```python
import pandas as pd
import pandas_datareader as web
from pandas.tseries.offsets import DateOffset
import numpy as np
from time import strptime
import plotly.express as px

# import models
import pmdarima as pm
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
# from fbprophet import Prophet

# saving the model
import joblib
```

```python
def model_forcast_plot(dataset_,columns_):
    
    dataset = dataset_[columns_].copy()
    
# find the best model order for arima or sarima
    best_model_order = pm.auto_arima(dataset, 
                                     start_p=1, start_q=1, max_p=3, max_q=5, 
                                     start_P=0, start_Q=0, max_P=3, max_Q=3, m=11, 
                                     seasonal=True, trace=True, 
                                     d=1, D=1, 
                                     error_action='warn', 
                                     suppress_warnings=True, 
                                     random_state = 3, 
                                     n_fits=11).fit(dataset)
    
# saving the model
    file_name = columns_ +'.sav'
    joblib.dump(best_model_order,file_name)
    
#     forcast = best_model_order.predict(n_periods=23, return_conf_int=False)
    
    
#     future_dates = [dataset.index[-1]+ DateOffset(months = x)for x in range(0,23)]
#     future_dates_df = pd.DataFrame(forcast,index=future_dates[0:])
#     future_dataset = pd.concat([dataset,future_dates_df],axis='columns')

#     future_dataset.columns = ['Actual','Forcast']
    
# # +++++++++++++++++++++++++++++ FORCAST PLOT +++++++++++++++++++++++++
#     print(columns_)
#     title_name = str(input('enter the title for the graph : '))
#     fig = px.line(x = future_dataset.index, y = future_dataset['Actual'],title=title_name)
#     fig.add_scatter(x=future_dataset.index, y = future_dataset['Forcast'])
#     fig.update_xaxes(
#         rangeslider_visible=True,
#         )
#     fig.show()
# #     plt.plot(future_dataset['Actual'])
# #     plt.plot(future_dataset['Forcast'])
# #     plt.title('gdp')
# #     plt.show()
```

# interest rate

```python
interest_rate = pd.read_csv('Interest Rates RBI .csv',header=1)
```

```python
interest_rate.dropna(inplace=True)
```

```python
month  = interest_rate.Month.str.split('-',expand=True)[0].to_list()
```

```python
for i in range(len(month)):
    if month[i][0:2] == 'Ja':
        month[i] = "January"
    
    elif month[i][0] == 'F':
        month[i] = "February"
    
    elif month[i][0:3] == 'Mar':
        month[i] = "March"
    
    elif month[i][0:2] == 'Ap':
        month[i] = "April"
    
    elif month[i][0:3] == 'May':
        month[i] = "May"
    
    elif month[i][0:3] == 'Jun':
        month[i] = "June"
    
    elif month[i][0:3] == 'Jul':
        month[i] = "July"
    
    elif month[i][0:2] == 'Au':
        month[i] = "August"
    
    elif month[i][0:3] == 'Sep':
        month[i] = "September"
    
    elif month[i][0] == 'O':
        month[i] = "October"
    
    elif month[i][0] == 'N':
        month[i] = "November"
    
    elif month[i][0] == 'D':
        month[i] = "December"
```

```python
interest_rate['Date'] = pd.to_datetime({'year':interest_rate.Month.str.split('-',expand=True)[1],
               'month':[strptime(i,'%B').tm_mon for i in month],
               'day':np.ones(interest_rate.shape[0])})
```

```python
interest_rate.drop(['Month'],axis='columns',inplace=True)
```

```python
interest_rate.sort_values(by=['Date'],ascending=False,inplace=True)
interest_rate.set_index('Date',inplace = True)
interest_rate.index = pd.to_datetime(interest_rate.index,format= "%b-%y") 
interest_rate.sort_index(ascending=True,inplace=True)
```

```python
interest_rate = interest_rate.loc[~interest_rate.index.duplicated(keep='first')]
```

```python
for i in interest_rate.columns:
    model_forcast_plot(interest_rate,i)
```
